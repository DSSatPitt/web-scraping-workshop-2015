```python
import os
import sys
import twitter
from twitter.oauth import write_token_file, read_token_file
from twitter.oauth_dance import oauth_dance
import json
from keys import APP_NAME, CONSUMER_KEY, CONSUMER_SECRET
```

```python
def login(app_name=APP_NAME,
                consumer_key=CONSUMER_KEY, 
                consumer_secret=CONSUMER_SECRET, 
                token_file='out/twitter.oauth'):

    try:
        (access_token, access_token_secret) = read_token_file(token_file)
    except IOError:
        (access_token, access_token_secret) = oauth_dance(app_name, consumer_key,
                consumer_secret)

        if not os.path.isdir('out'):
            os.mkdir('out')

        write_token_file(token_file, access_token, access_token_secret)

        print("OAuth Success. Token file stored to", token_file, file=sys.stderr)

    return twitter.Twitter(auth=twitter.oauth.OAuth(access_token, access_token_secret,
                           consumer_key, consumer_secret))

def login_stream(app_name=APP_NAME,
                consumer_key=CONSUMER_KEY, 
                consumer_secret=CONSUMER_SECRET, 
                token_file='out/twitter.oauth'):
    try:
        (access_token, access_token_secret) = read_token_file(token_file)
    except IOError:
        (access_token, access_token_secret) = oauth_dance(app_name, consumer_key,
                consumer_secret)

        if not os.path.isdir('out'):
            os.mkdir('out')

        write_token_file(token_file, access_token, access_token_secret)

        print("OAuth Success. Token file stored to", token_file, file=sys.stderr)

    return twitter.TwitterStream(auth=twitter.oauth.OAuth(access_token, access_token_secret,
                           consumer_key, consumer_secret))

```

```python
t = login(APP_NAME, CONSUMER_KEY,CONSUMER_SECRET)
```

```python
t.statuses.user_timeline(screen_name="mcburton")
```

```python
iconf_tweets = t.search.tweets(q="#iconf15", count=1000)
```

```python
with open("iconf-tweets.json", 'a') as f:
    for tweet in iconf_tweets['statuses']:
        print("{0}: {1}\n".format(tweet['user']['screen_name'],tweet['text']))
        json.dump(tweet, f)
        f.write("\n")
```

```python
streamy = login_stream(APP_NAME, CONSUMER_KEY,CONSUMER_SECRET)
```

```python
with open("iconf-tweets.json", 'a') as f:

    try:
        for tweet in streamy.statuses.filter(track="#iconf15"):
            print("{0}: {1}\n".format(tweet['user']['screen_name'],tweet['text']))
            json.dump(tweet, f)
            f.write("\n")
    except(KeyboardInterrupt):
        print("Closing the stream")
```
